```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("data/covid-19-municipios-sp.csv")
```

```python
df.Dia =  pd.to_datetime(df.Dia, format='%d-%m-%Y')
```

```python
df.head()
```

## Mapa Base

```python
def generateBaseMap(default_location=[-22.292690, -48.558171], default_zoom_start=7):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
base_map
```

# Mapa de Calor

```python
from folium.plugins import HeatMap

map_confirmados = generateBaseMap()
HeatMap(data=df[['Latitude', 'Longitude', 'Confirmados']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(map_confirmados)
map_confirmados
```

```python
# Epicentros
map_confirmados.add_child(folium.ClickForMarker(popup='Epicentro-São Paulo-SP'))
```

# Adicionando Controlador de Tempo

```python
df_day_list = []
for day in df.Dia.sort_values().unique():
    df_day_list.append(df.loc[df.Dia == day, ['Latitude', 'Longitude', 'Confirmados']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist())
```

```python
len(df_day_list)
```

```python
df_day_list
```

```python
from folium.plugins import HeatMapWithTime

casos_confirmados_map = generateBaseMap()
casos_covid_time = HeatMapWithTime(data=df_day_list, radius=5, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True)
casos_covid_time.add_to(casos_confirmados_map)
casos_confirmados_map
```

```python
casos_confirmados_map.save("covid-casos-confirmados-map-sp.html")
```
